```python
import pandas as pd
import numpy as np

```

```python
path = '/pscratch/sd/f/fkiwit/work/result_compression_20240916_135045.csv'

df = pd.read_csv(path)
print(df)
```

```python
import matplotlib.pyplot as plt

# Assuming your DataFrame is called df

# Filter the DataFrame for number_of_nodes equal to 1 and 2
df_nodes_1 = df[df['number_of_nodes'] == 1]
df_nodes_2 = df[df['number_of_nodes'] == 2]

# Plot compute_time_sec as a function of cores_per_node for number_of_nodes == 1
plt.plot(df_nodes_1['cores_per_node'], df_nodes_1['compute_time_sec'], label='1 Node', marker='o')

# Plot compute_time_sec as a function of cores_per_node for number_of_nodes == 2
plt.plot(df_nodes_2['cores_per_node'], df_nodes_2['compute_time_sec'], label='2 Nodes', marker='o')

# Add labels and title
plt.xlabel('Cores per Node')
plt.ylabel('Compute Time (sec)')
plt.title('Compute Time vs. Cores per Node for Different Node Counts')
plt.legend()

# Show the plot
plt.show()
```

```python
# Plot start_ray_time_sec as a function of cores_per_node for number_of_nodes == 1
plt.plot(df_nodes_1['cores_per_node'], df_nodes_1['start_ray_time_sec'], label='1 Node', marker='o')

# Plot start_ray_time_sec as a function of cores_per_node for number_of_nodes == 2
plt.plot(df_nodes_2['cores_per_node'], df_nodes_2['start_ray_time_sec'], label='2 Nodes', marker='o')

# Add labels and title
plt.xlabel('Cores per Node')
plt.ylabel('Start Time (sec)')
plt.title('Start Time vs. Cores per Node for Different Node Counts')
plt.legend()

# Show the plot
plt.show()
```

```python
# Plot load_data_time_sec as a function of cores_per_node for number_of_nodes == 1
plt.plot(df_nodes_1['cores_per_node'], df_nodes_1['load_data_time_sec'], label='1 Node', marker='o')

# Plot load_data_time_sec as a function of cores_per_node for number_of_nodes == 2
plt.plot(df_nodes_2['cores_per_node'], df_nodes_2['load_data_time_sec'], label='2 Nodes', marker='o')

# Add labels and title
plt.xlabel('Cores per Node')
plt.ylabel('Load Time (sec)')
plt.title('Load Time vs. Cores per Node for Different Node Counts')
plt.legend()

# Show the plot
plt.show()
```
